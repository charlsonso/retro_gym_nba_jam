```python
import retro
from IPython import display
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

players = 2
env = retro.make('NbaJamTE-Snes',players=players)
obs = env.reset()
env.action_space
```

```python
#Sampling for 1 player
env.action_space.sample()[:int(env.action_space.shape[0]/players)]

#Two independent players
player1_input = env.action_space.sample()[:int(env.action_space.shape[0]/players)]
player2_input = env.action_space.sample()[:int(env.action_space.shape[0]/players)]

both_player_input = np.hstack([player1_input, player2_input])

print(both_player_input)
```

```python
#display
import numpy as np
img = plt.imshow(env.render(mode='rgb_array')) # only call this once
for _ in range(100):
    
    img.set_data(env.render(mode='rgb_array')) # just update the data
    display.display(plt.gcf())
    display.clear_output(wait=True)
    action = env.action_space.sample()
    
    player1_input = env.action_space.sample()[:int(env.action_space.shape[0]/players)]
    player2_input = env.action_space.sample()[:int(env.action_space.shape[0]/players)]

    both_player_input = np.hstack([player1_input, player2_input])
    env.step(both_player_input)
```
